```
import numpy as np
import tensorflow as tf
import pandas as pd
```

```
tf.__version__
```

```
from google.colab import files
uploaded = files.upload()
```

```
import io
train_data = pd.read_csv(io.BytesIO(uploaded['Churn_Modelling.csv']))
```

```
train_data
```

```
y = train_data['Exited']
```

```
train_data.drop('RowNumber', axis=1, inplace=True)
```

```
train_data.drop('CustomerId', axis=1, inplace=True)
train_data.drop('Surname', axis=1, inplace=True)
train_data.drop('Exited', axis=1, inplace=True)

```

```
x = train_data
```

```
x.info()
```

```
x = pd.get_dummies(x)
```

```
x
```

```
x.drop('Geography_France', axis=1,inplace=True)
```

```
x.drop('Gender_Female', axis=1,inplace=True)
```

```
from sklearn.model_selection import train_test_split
x_train, x_eval, y_train, y_eval = train_test_split(x,y,test_size=0.2,random_state=0)
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_eval = sc.fit_transform(x_eval)
```

```
x_train
```

```
ann = tf.keras.models.Sequential()
```

```
ann.add(tf.keras.layers.Dense(units=11, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```
ann.add(tf.keras.layers.Dense(units=1, activation='tanh'))
```

```
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```
ann.fit(x_train,y_train,batch_size=32,epochs=100)
```

```
X = [[600,40,3,60000,2,1,1,50000,0,0,1]]

sc.transform(X)
```

```
print((ann.predict(X))>0.5)
```

```
y_pred = ann.predict(x_eval)
```

```
y_pred
```

```
l = []
for pred in y_pred:
  if pred>0.5:
    l.append(1)
  else:
    l.append(0)  

```

```
l
```

```
y_eval = np.array(y_eval)
```

```
y_eval
```

```
l = np.array(l)
```

```
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_eval,l)
print(cm)
accuracy_score(y_eval,l)
```
